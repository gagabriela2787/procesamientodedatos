```python
import pandas as pd
import matplotlib.pyplot as plt
# Cargar el archivo en un DataFrame
df = pd.read_csv('historico-nombres.csv')

# Mostrar las primeras 5 filas del DataFrame
print(df.head())
```

```python
# Consulta: Mostrar la cantidad total de registros en el DataFrame
total_registros = len(df)
print("Cantidad total de registros:", total_registros)

# Consulta: Calcular la cantidad total de nombres Ãºnicos
nombres_unicos = df['nombre'].nunique()
print("Cantidad total de nombres Ãºnicos:", nombres_unicos)

# Consulta: Calcular la cantidad total de registros por aÃ±o
registros_por_anio = df.groupby('anio')['cantidad'].sum()

# GrÃ¡fico: Visualizar la cantidad de registros por aÃ±o
plt.figure(figsize=(10, 6))
registros_por_anio.plot(kind='line')
plt.title('Cantidad de Registros por AÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('Cantidad')
plt.show()
```
